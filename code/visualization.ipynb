### 2 visualization

#### 2.1 Average Delay Time by Airline in 2023 Holiday season

```python
data = filtered_df[filtered_df['DepTime']>= '2023-11-01']
```

```python
data.head()
```

```python
data['delay'] = (data['CRSArrTime_num'] - data['ArrTime_num'])/60
```

```python
from matplotlib.patches import Patch
avg_delay_by_airline = {}

for col in df.columns[-11:-1]: 
    airline_data = data[data[col] == 1]

    avg_delay = airline_data['delay'].mean()
    avg_cancelled_rate = airline_data['Cancelled'].mean() * 100 
    avg_delay_by_airline[col] = {'Average_Delay': avg_delay, 'Average_Cancelled_Rate': avg_cancelled_rate}


avg_delay_df = pd.DataFrame.from_dict(avg_delay_by_airline, orient='index').reset_index()
avg_delay_df = avg_delay_df.rename(columns={'index': 'Airline'})


avg_delay_df = avg_delay_df.sort_values(by='Average_Delay', ascending=False)

colors = {
    'Airline_AS': '#002240',
    'Airline_HA': '#003366',
    'Airline_UA': '#004C99',
    'Airline_WN': '#0066CC',
    'Airline_AA': '#3399FF',
    'Airline_NK': '#66B2FF',
    'Airline_B6': '#99CCFF',
    'Airline_F9': '#CCE6FF',
    'Airline_G4': '#E6F7FF',
    'Airline_DL': '#B0E0E6'
}
plt.figure(figsize=(10, 6))
sns.barplot(data=avg_delay_df, x="Airline", y="Average_Delay", palette=colors)
plt.title("Average Delay Time by Airline in 2023 Holiday season")
plt.xlabel("Airline")
plt.ylabel("Average Delay (minutes)")

legend_labels = [Patch(color=color, label=airline) for airline, color in colors.items()]
plt.legend(handles=legend_labels, title="Airline")
plt.savefig('Average Delay Time by Airline in 2023 Holiday season.png', dpi=300)
plt.show()
```

```python
avg_delay_by_airline = {}

for col in df.columns[-11:-1]: 
    avg_delay_by_airline[col] = data.loc[data[col] == 1, 'Cancelled'].mean()*100

avg_delay_df = pd.DataFrame(list(avg_delay_by_airline.items()), columns=['Airline', 'Cancelled_rate'])
avg_delay_df = avg_delay_df.sort_values(by='Cancelled_rate', ascending=False)

colors = {
    'Airline_AS': '#002240',
    'Airline_HA': '#003366',
    'Airline_UA': '#004C99',
    'Airline_WN': '#0066CC',
    'Airline_AA': '#3399FF',
    'Airline_NK': '#66B2FF',
    'Airline_B6': '#99CCFF',
    'Airline_F9': '#CCE6FF',
    'Airline_G4': '#E6F7FF',
    'Airline_DL': '#B0E0E6'
}

plt.figure(figsize=(10, 6))
sns.barplot(data=avg_delay_df, x="Airline", y="Cancelled_rate", palette=colors)
plt.title("Cancelled Rate by Airline in 2023 Holiday season")
plt.xlabel("Airline")
plt.ylabel("Cancelled Rate(%)")

legend_labels = [Patch(color=color, label=airline) for airline, color in colors.items()]
plt.legend(handles=legend_labels, title="Airline")
plt.savefig('Cancelled Rate by Airline in 2023 Holiday season.png', dpi=300)
plt.show()
```

#### 2.2 Cancelled rate by day of week in 2023 Holiday season (DepTime)

```python
data['Hour'] =data['DepTime'].dt.hour  
data['DayOfWeek'] = data['DepTime'].dt.day_name() 

day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
data['DayOfWeek'] = pd.Categorical(data['DayOfWeek'], categories=day_order, ordered=True)

cancel_by_hour = data.groupby(['Hour', 'DayOfWeek'])['Cancelled'].mean().reset_index()
```

```python
cancel_by_hour
```

```python
sns.lineplot(data=cancel_by_hour, x='Hour', y='Cancelled', hue='DayOfWeek', marker="o")

plt.xlim(-1, 24)
plt.title('Cancelled Rate by Scheduled Departure Time during 2023 Holiday Season')
plt.xlabel('Scheduled Departure Time(Hour)')
plt.ylabel('Cancellation Rate')
plt.legend(title='Day of Week')
plt.savefig('Cancelled Rate by Scheduled Departure Time during 2023 Holiday Season.png', dpi=300)
plt.show()
```

#### 2.3 Average Delay Time by day of week in 2023 Holiday season (DepTime and ArrTime)

```python
data['Hour'] =data['DepTime'].dt.hour 
data['DayOfWeek'] = data['DepTime'].dt.day_name() 
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
data['DayOfWeek'] = pd.Categorical(data['DayOfWeek'], categories=day_order, ordered=True)

delay_by_hour = data.groupby(['Hour', 'DayOfWeek'])['delay'].mean().reset_index()
```

```python
sns.lineplot(data=delay_by_hour, x='Hour', y='delay', hue='DayOfWeek', marker="o")

plt.xlim(-1, 24)
plt.title('Average Delay Time by Scheduled Departure Time during 2023 Holiday Season')
plt.xlabel('Scheduled Departure Time (Hour)')
plt.ylabel("Average Delay (minutes)")
plt.legend(title='Day of Week')
plt.savefig('Average Delay Time by Scheduled Departure Time during 2023 Holiday Season.png', dpi=300)
plt.show()
```
